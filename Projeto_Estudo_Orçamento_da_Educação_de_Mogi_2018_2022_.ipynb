```python
def ColunasEdu (dataset, num):
  x = 0
  lista = []
  y = 0
  while x <= 9:
    y = dataset.values[num]
    lista.append(y[1:])
    num += 1
    x = x + 1
  return lista

def criardb (dataset, ColunasEdu, ano):
  array2 = dataset.values[16]
  array2 = np.insert(array2, 12, 'ANO')
  array3 = ['ENSINO FUNDAMENTAL','ENSINO MEDIO','ENSINO PROFISSIONAL',
            'ENSINO SUPERIOR','ENSINO INFANTIL','ENSINO JOVENS E ADULTOS',
            'ENSINO ESPECIAL','ENSINO BASICO','FU12 - ADMINISTRACAO','FU12 - DEMAIS SUBFUNCOES']
  values = []
  for sublista in ColunasEdu:
    sublista = np.insert(sublista, 11, str(ano))
    values.append(sublista)
  dici = dict(zip(array3, values))
  df = pd.DataFrame(dici, index=array2[1:])
  return df

#essa função vai me retornar as colunas no lugar das linhas e vice e versa, fiz desse jeito pois usando um .T
#depois, deixo o script mais fácil de ser retrabalhado do que criar uma função para cada linha, então decidi
#retornar uma lista de arrays que serão os valores (11 arrays) setando 11 indices (array2), pois dessa forma
#consigo criar uma função com while para acelerar novos reprocessos
```

```python

class filtragem:
  def __init__ (self, base_dados, palavra, lista_vazia):
    self.base_dados = base_dados
    self.palavra = palavra
    self.lista_vazia = lista_vazia

  def filtro_categoria (self):
    palavra_filtro = self.palavra
    lista_colunas = self.base_dados.columns
    colunas_filtradas = [self.lista_vazia.append(x) for x in lista_colunas if palavra_filtro in x]
```

```python
df_copia = grupamento.copy()
```

```python
#vamos normalizar meus dados em uma para poder analisar sua distribuição com mais efetividade
#também para poder fazermos uma regressão linear depois (mesmo que a quantidade de dados não nos ajude e o resultado
#do teste não será boa)

for column in df_copia.columns:
    df_copia[column] = df_copia[column]  / df_copia[column].abs().max()

```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
import scipy.stats as ss
from sklearn.model_selection import train_test_split
```

```python
X = df_copia['DOTAÇÃO INICIAL ']
Y = df_copia['DESPESAS EMPENHADAS NO BIMESTRE ']
```

```python
 r = ss.pearsonr(X,Y)
 print(r)
```

```python
#CORRELAÇÃO TENDE A 0 - SEM CORRELAÇÃO - isso contraria o heatmap acima (talvez a normalizacão)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, train_size=0.7, test_size=0.3, shuffle=False
)

reg = LinearRegression().fit(X_train.values.reshape(-1, 1), Y_train)
```

```python
from sklearn.metrics import r2_score

y_pred = reg.predict(X_test.values.reshape(-1,1))

print(mean_squared_error(Y_test, y_pred))
print(mean_absolute_error(Y_test, y_pred))
print(r2_score(Y_test, y_pred))
```
